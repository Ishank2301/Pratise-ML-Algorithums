# Decision Tree Regression

## Importing the libraries

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Importing the dataset

```
df = pd.read_csv('Position_Salaries.csv')

X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```
print(X)
```

```
print(y)
```

## Training the Decision Tree Regression model on the whole dataset

```
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(random_state=0)
regr.fit(X,y)
y_pred = regr.predict([[6.5]])
```

## Predicting a new result

*   If we don't apply the Higher resolution formula
*   The real Decision Graph graph is below{This graph is used to explain drawbacks}



```
plt.scatter(X,y,color="Blue",label="Data")
plt.plot(X,regr.predict(X),color="Blue",label="Model_Prediction")
plt.title("Decision Tree Regression",color="Green",size=10)
plt.xlabel("Positions",color="Green",size=10)
plt.ylabel("Salaries",color="Green",size=10)
plt.tick_params(axis="both",    # apply to major ticks
    direction='out',    # ticks point inward
    color='red',       # tick color is red
    labelsize=12,      # label font size
    labelbottom=True )
plt.legend()
plt.show()
```

## Visualising the Decision Tree Regression results (higher resolution)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X,y,color="Red",label="Data")
plt.plot(X_grid,regr.predict(X_grid),color="Blue",label="Model_Prediction")
plt.title("Decision Tree Regression",color="Green",size=10)
plt.xlabel("Positions",color="Green",size=10)
plt.ylabel("Salaries",color="Green",size=10)
plt.tick_params(axis="both",    # apply to major ticks
    direction='out',    # ticks point inward
    color='red',       # tick color is red
    colors='Purple',       # tick color is red
    labelsize=12,      # label font size
    labelbottom=True )
plt.legend()
plt.show()
```
