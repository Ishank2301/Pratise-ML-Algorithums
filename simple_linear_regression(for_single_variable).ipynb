# Simple Linear Regression

## Importing the libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Importing the dataset

```python
df=pd.read_csv('Salary_Data.csv')
df.info()
df.describe()
df.head(10)
df.isnull().sum()
```

# Preprocess the data


```python
X = df.iloc[:,:-1].values
print(X)
```

```python
y = df.iloc[:,-1].values
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)  # y_pred(predicted salary) because y_test contains the real salary:
```

## Visualising the Training set results

```python
plt.scatter(X_train,y_train,color="Red",label="Train Data")
plt.plot(X_train,regressor.predict(X_train),color="Black",label="Prediction")
plt.title("Salary vs Experience(Training set)")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test,y_test,color="Red",label="Test Data")
plt.plot(X_train,regressor.predict(X_train),color="Black",label="Prediction")
plt.title("Salary vs Experience(Test set)")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

# Making a single prediction (for example the salary of an employee with 12 years of experience)

```python
print(regressor.predict([[12]]))
```

Therefore, our model predicts that the salary of an employee with 12 years of experience is $138371.45.

Important Note : Notice that the value of the feature 12 years was input ina double pair of square brackets. That's beacause the "predict" method always expects a 2D array as the format o the inputs.

12 -> scalar
[12] -> 1D array
[[12]] -> 2D array

# Getting the final linear regression eqaution with the values of the coefficient

```python
print(regressor.coef_)
print(regressor.intercept_)
```

Therefore, the equation of our simple linear regression model is:

Salary=9345.94×YearsExperience+26816.19

Important Note: To get these coefficients we called the "coef_" and "intercept_" attributes from our regressor object. Attributes in Python are different than methods and usually return a simple value or an array of values.
